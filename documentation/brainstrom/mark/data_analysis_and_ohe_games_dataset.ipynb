# One-Hot Encoding Clean Games Dataset

The Games dataset has been cleaned in 'data_analysis_and_cleaning_games_dataset.ipynb'. 

Two of the columns in the clean Games contain categorical data (Genre and Team), one of the objectives is to test whether one-hot encoding these columns and including them in the Decision Trees and Random Forests produce better results than excluding them altogether.

```python
import pandas as pd
```

```python
# create a pandas DataFrame called "games" containing the clean data

games = pd.read_csv('C:/Users/Mark/Documents/repos/group-coursework-ha/data/games_clean.csv')
```

```python
# display the top five rows of the DataFrame to confirm the clean data has been imported properly and is ready to be one-hot encoded

games.head() 
```

# One-Hot Encoding

```python
# One-Hot Encode Team and Genres columns

# Team and Genres columns are nominal, so use One-Hot Encoding with the get_dummies() function

temp_df = pd.get_dummies(games[['Team','Genres']])

# merge one-hot encoded columns back with original DataFrame

games = games.join(temp_df)

# drop the original Team and Genres columns as they're no longer required

games.drop('Team', axis=1, inplace=True)
games.drop('Genres', axis=1, inplace=True)
```
