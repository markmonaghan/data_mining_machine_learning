```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor  # Import Decision Tree Regressor 
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.metrics import r2_score
```

```python
# create a pandas dataframe from the data in GitHub

games = pd.read_csv("https://raw.githubusercontent.com/dmml-heriot-watt/group-coursework-ha/mark-branch/data/games_clean.csv?token=GHSAT0AAAAAACJ3NTU5ZWT4VSHAIT2KWM7SZKG2UPA")
```

```python
games.head()
```

```python
# split dataset into features and target variable

# remove 'Title' from features as its only used for context on which game is being referred to
# remove 'Rating' as thats the target variable

feature_columns = ['Number of Reviews','Plays','Playing','Active Users']
```

```python
X = games[feature_columns] # features

y = games.Rating # target variable
```

# Model Testing - Splitting the data - Holdout Method

```python
# split dataset into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70% training and 30% test
```

# Build Decision Tree Regression Models

```python
# create Decision Tree regressor objects

dtr = DecisionTreeRegressor(max_depth=3, random_state=42)

# train Decision Tree regressor

dtr = dtr.fit(X_train,y_train)
```
